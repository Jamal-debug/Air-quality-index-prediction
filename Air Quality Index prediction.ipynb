```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
df=pd.read_csv('AirQualityUCI.csv')
```

```python
df.head()
```

```python
df.info()
```

```python
df.drop(['Unnamed: 15','Unnamed: 16'],axis=1 ,inplace=True)
```

```python
df.corr()
```

```python
for x in df:
    print(x)
    print(df[x].isnull().value_counts())
```

```python
df.drop(['Date','Time'],axis=1,inplace=True)
```

```python
df.dropna(subset=['RH'],axis=0,inplace=True)

df.dtypes
```

```python
sns.pairplot(data=df)
```

```python
from scipy import stats
zscore=np.abs(stats.zscore(df))

thresh=3
print(np.where(zscore>3))
print(zscore[15][1])
cdf=df
cdf=cdf[(zscore<3).all(axis=1)]
```

```python
x=cdf[['PT08.S1(CO)','C6H6(GT)','PT08.S2(NMHC)','PT08.S3(NOx)','PT08.S4(NO2)','PT08.S5(O3)','T']]
y=cdf['RH']
```

```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=.20,random_state=0)
```

```python
from sklearn.ensemble import RandomForestRegressor           #import random forest regressor
rf=RandomForestRegressor()
rf.fit(xtrain,ytrain)
yhat=rf.predict(xtest)

from sklearn.metrics import r2_score,mean_squared_error,mean_absolute_error
print("RMSE : ",np.sqrt(mean_squared_error(ytest,yhat)))
print(r2_score(ytest,yhat))
print(df.shape)
```
